```python
# poly.py
# -*- coding: utf-8 -*-
from __future__ import unicode_literals
import os
from pandas_datareader import data
import pandas as pd
import numpy as np
import sklearn.pipeline as pl
import sklearn.preprocessing as sp
import sklearn.linear_model as lm
import sklearn.metrics as sm
import matplotlib.pyplot as mp
import matplotlib.dates as md
```

```python
symbol = str(input('請輸入股票代號：'))
data_source = 'yahoo'
start_date = str(input('請輸入開始日期：'))
end_date = str(input('請輸入結束日期：'))
n = int(input("請輸入？次多項式迴歸："))
```

```python
df = data.DataReader(symbol, data_source, start_date, end_date)
df.to_csv(os.path.join('./csvFile', 'stock_'+symbol+'.csv'))
df = pd.read_csv(os.path.join('./csvFile', 'stock_'+symbol+'.csv'))
df['Date'] = pd.to_datetime(df['Date'])
print(df.dtypes)
```

```python
x = np.zeros(df['Close'].size)
for i in range(df['Close'].size):
    x[i] = float(i)
x = x.reshape(x.size, 1)
y = np.array([df['Close']])[-1]
print(x.shape,y.shape)
```

```python
model = pl.make_pipeline(sp.PolynomialFeatures(n), lm.LinearRegression())
model.fit(x, y)
pred_y = model.predict(x)
print(sm.r2_score(y, pred_y))
```

```python
mp.figure(symbol+' '+'Polynomial Regression', facecolor='lightgray')
mp.title(symbol+'\n''Polynomial Regression', fontsize=20)
mp.rcParams['figure.figsize'] = (20.0, 10.0) 
mp.xlabel('Date', fontsize=14)
mp.ylabel('Close', fontsize=14)
ax = mp.gca()
# 設置水平坐標每三個月為主刻度
ax.xaxis.set_major_locator(md.MonthLocator(range(1, 12, 3)))
# 設置水平坐標每一個月為次刻度
ax.xaxis.set_minor_locator(md.MonthLocator())
# 設置水平坐標主刻度標籤格式
ax.xaxis.set_major_formatter(md.DateFormatter('%d %b %Y'))
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
mp.plot(df['Date'], df['Close'], c='lightgray', label='Close Price')
mp.plot(df['Date'], pred_y,  c='dodgerblue', label='Regression')
mp.legend()
mp.gcf().autofmt_xdate()
mp.show()
```
