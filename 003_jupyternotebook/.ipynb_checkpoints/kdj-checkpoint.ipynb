```python
# ebb.py
# -*- coding: utf-8 -*-
from __future__ import unicode_literals
import os
from pandas_datareader import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import matplotlib.dates as md
```

```python
symbol = str(input('請輸入股票代號：'))
data_source = 'yahoo'
start_date = str(input('請輸入開始日期：'))
end_date = str(input('請輸入結束日期：'))
rsv_days = int(input('請輸入RSV的參數：'))
```

```python
df = data.DataReader(symbol, data_source, start_date, end_date)
df.to_csv(os.path.join('./csvFile', 'stock_'+symbol+'.csv'))
df = pd.read_csv(os.path.join('./csvFile', 'stock_'+symbol+'.csv'))
df['Date'] = pd.to_datetime(df['Date'])
print(df)
print(df.dtypes)
```

```python
# 未成熟隨機值RSV
rsvs = np.zeros(df['Close'].size-(rsv_days-1))
for i in range(rsvs.size):
    rsvs[i] = (df['Close'][::-1][i+(rsv_days-1):i+rsv_days]-df['Close'][::-1][i:i+rsv_days].min())\
                /(df['Close'][::-1][i:i+rsv_days].max()-df['Close'][::-1][i:i+rsv_days].min())\
                *100
# print(rsvs.size)
# K值
ks = np.zeros(df['Close'].size-(rsv_days-2))
for i in range(ks.size):
    ks[i] = 50 if i == 0 else 2/3*ks[i-1] + 1/3*rsvs[i-1]
# print(ks.size)
# D值
ds = np.zeros(df['Close'].size-(rsv_days-2))
for i in range(ds.size):
    ds[i] = 50 if i == 0 else 2/3*ds[i-1] + 1/3*ks[i]
# print(ds.size)
# J值
js = np.zeros(df['Close'].size-(rsv_days-2))
for i in range(js.size):
    js[i] = 3*ds[i] - 2*ks[i]
# print(ds.size)
```

```python
mp.figure(symbol+' '+'Stochastics　KDJ　Line', facecolor='lightgray')
mp.title(symbol+'\n'+'Stochastics　KDJ　Line', fontsize=20)
mp.rcParams['figure.figsize'] = (20.0, 10.0)
mp.xlabel('Date', fontsize=14)
mp.ylabel('Close', fontsize=14)
ax = mp.gca()
# 設置水平坐標每三個月為主刻度
ax.xaxis.set_major_locator(md.MonthLocator(range(1, 12, 3)))
# 設置水平坐標每一個月為次刻度
ax.xaxis.set_minor_locator(md.MonthLocator())
# 設置水平坐標主刻度標籤格式
ax.xaxis.set_major_formatter(md.DateFormatter('%d %b %Y'))
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')    
mp.plot(df['Date'][rsv_days-2:], ks, c='orangered', label='K'+str(rsv_days))
mp.plot(df['Date'][rsv_days-2:], ds, c='limegreen', label='D'+str(rsv_days))
mp.plot(df['Date'][rsv_days-2:], js, c='dodgerblue', label='J'+str(rsv_days))
mp.legend()
mp.gcf().autofmt_xdate()
mp.show()
```
