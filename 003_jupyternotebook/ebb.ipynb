```python
# ebb.py
# -*- coding: utf-8 -*-
from __future__ import unicode_literals
import os
from pandas_datareader import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import matplotlib.dates as md
```

```python
symbol = str(input('請輸入股票代號：'))
data_source = 'yahoo'
start_date = str(input('請輸入開始日期：'))
end_date = str(input('請輸入結束日期：'))
ema_days = int(input('請輸入指數加權移動平均線的天數：'))
stds_n = float(input('請輸入均線上下？個標準差：'))
```

```python
df = data.DataReader(symbol, data_source, start_date, end_date)
df.to_csv(os.path.join('./csvFile', 'stock_'+symbol+'.csv'))
df = pd.read_csv(os.path.join('./csvFile', 'stock_'+symbol+'.csv'))
df['Date'] = pd.to_datetime(df['Date'])
print(df)
print(df.dtypes)
```

```python
weights = np.exp(np.linspace(-1, 0, ema_days))
weights /= weights.sum()
medios = np.convolve(df['Close'], weights[::-1], 'valid')
stds = np.zeros(medios.size)
for i in range(stds.size):
    stds[i] = df['Close'][i:i+ema_days].std()
lowers = medios - stds*stds_n
uppers = medios + stds*stds_n
```

```python
mp.figure(symbol+' '+'Exponential Moving Average Bollinger Bands', facecolor='lightgray')
mp.title(symbol+'\n'+'Exponential Moving Average Bollinger Bands', fontsize=20)
mp.rcParams['figure.figsize'] = (20.0, 10.0) 
mp.xlabel('Date', fontsize=14)
mp.ylabel('Close', fontsize=14)
ax = mp.gca()
# 設置水平坐標每三個月為主刻度
ax.xaxis.set_major_locator(md.MonthLocator(range(1, 12, 3)))
# 設置水平坐標每一個月為次刻度
ax.xaxis.set_minor_locator(md.MonthLocator())
# 設置水平坐標主刻度標籤格式
ax.xaxis.set_major_formatter(md.DateFormatter('%d %b %Y'))
mp.tick_params(labelsize=10)
mp.grid(linestyle=':')
mp.plot(df['Date'], df['Close'], c='lightgray', label='Close Price')
mp.plot(df['Date'][ema_days-1:], medios, c='dodgerblue', label='Medio'+'(EMA'+str(ema_days)+')')
mp.plot(df['Date'][ema_days-1:], lowers, c='limegreen', label='Lower')
mp.plot(df['Date'][ema_days-1:], uppers, c='orangered', label='Upper')
mp.legend()
mp.gcf().autofmt_xdate()
mp.show()
```
